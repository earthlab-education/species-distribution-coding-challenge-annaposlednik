## *Pandion haliaetus* 

Osprey (*Pandion haliaetus*) are charismatic raptors that are most known for their diving ability, as their diet is largely composed of live fish. Osprey migrate annually in the spring and fall between North America for breeding and the Carribbean and South America for the winter. These large hawks have a life-span of 15-20 years and log upwards of 160,000 miles in their lifetime. 

Osprey's diet of live fish, makes these birds of prey definitive hosts for a number of different marine parasites, including the trematode *Scaphanocephallus sp.* which causes black spot syndrome in Carribbean fish. This syndrome is emerging across fish species and islands of the Carribbean, and may pose implications for coral reef ecosystems, as this parasite is known to infect a number of important herbivorous fish that feed on algae. One hypothesis for why scientists are seeing a recent emergence of this parasite in the Carribbean could be linked to osprey populations rebounding following the ban on DDT, allowing for greater spread of the parasite through osprey migrations. To better understand the possible extent of the spread, range, and transmission of this parasite, I chose to map the migrations of osprey. 

## Data Description

Data on osprey observations were taken from the <a href="https://www.gbif.org/what-is-gbif/">Global Biodiversity Information Facility</a> (GBIF), which is an international network that provides free and open access biodiversity data about all types of life on Earth. This network provides data collected from a variety of sources including museum specimens, DNA barcoding, and smartphone photos. 

For this analysis data on ospreys (*Pandion haliaetus*) migration for the year 2023 was downloaded from GBIF.

 <a href="https://www.geographyrealm.com/terrestrial-ecoregions-gis-data/"> Ecoregion data </a> was used from <a href="https://developers.google.com/earth-engine/datasets/catalog/RESOLVE_ECOREGIONS_2017"> RESOLVE Ecoregions dataset</a>. All ecoregions were included.



## Methods Description

Osprey occurrence data was filtered to include all occurrences from 2023 with coordinates available, and downloaded from GBIF. This dataframe, which included month and coordinates as geometry, was then converted to a GeoDataFrame using geopandas using the WGS84 projection as the coordinate reference system. Then, ecoregion data from RESOLVE was spatially joined with the osprey occurrence data for each month and location name, matching the WGS84 CRS. Monthly observations were counted for each region and then normalized by space and time. This involved taking mean occurrence for each month and ecoregion and diving the counts by both the mean occurrence per month and mean occurrence per region, to take into account the biases of crowd-sourced data (given we expect to see greater occurrences in locations and times with higher human activity). 
The normalized data was then plotted using hvplot, producing an interactive plot that depicts monthly migration patterns of osprey across 2023.
